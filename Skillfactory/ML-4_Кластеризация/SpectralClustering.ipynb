### Спектральная кластеризация Spectral Clustering

```python
# импорт бибилиотек
import pandas as pd
import seaborn as sns
# вызываем из sklearn SpectralClustering 
from sklearn.cluster import SpectralClustering
```

```python
# load data
df = pd.read_csv('data/pokemon.csv')
# произведем кластеризацию покемонов по уровню атаки и защиты
X = df[['Attack', 'Defense']]

# визуализируем через точечную диаграмму по двум признакам
sns.scatterplot(x=X.Attack, y=X.Defense)
```

```python
# запускаем кластеризацию, задав количество кластеров, равное 3 
spectral_clustering = SpectralClustering(n_clusters=3, random_state=42)

spectral_clustering.fit(X)
# делаем предсказание по кластеризации
#sc_predict = spectral_clustering.predict(X)

# получаем результаты кластеризации
spectral_predictions = spectral_clustering.labels_
```

```python
# визуализируем кластеризацию на 3 кластера
sns.scatterplot(x=df.Attack, y=df.Defense, c=spectral_predictions)
```
