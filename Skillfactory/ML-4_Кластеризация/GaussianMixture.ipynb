### EM-алгоритмы кластеризации

```python
# импортируем бибилиотеки
import numpy as np
from sklearn.mixture import GaussianMixture
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans # библиотека для кластеризации
```

```python
# load data
X = pd.read_csv('data/housing.csv')
X.head()
```

```python
# визуализируем через точечную диаграмму по двум признакам
sns.scatterplot(x=X.year, y=X.price)
```

```python
# создаем объект GaussianMixture
gm_clastering = GaussianMixture(n_components=3, random_state=42)

# обучаем модель
gm_clastering.fit(X)

# делаем предсказание по кластеризации
gm_predict = gm_clastering.predict(X)
```

```python
gm_predict
```

```python
# Далее можно визуализировать результаты, как мы делали с алгоритмом k-means
sns.scatterplot(x=X.year, y=X.price, c=gm_predict)
```

### Предсказание через KMeans

```python
# инициализация алгоритма KMeans
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10, random_state=42) # для 3 кластеров

# обучим модель через fit
kmeans.fit(X)

# произведем предсказание кластеризации
predict = kmeans.predict(X)
```

```python
# Далее можно визуализировать результаты, как мы делали с алгоритмом k-means
sns.scatterplot(x=X.year, y=X.price, c=predict)
```

### GMM справился с pадачей кластеризации лучше, чем KMeans
