### Кластеризация на основе плотности DBSCAN

```python
# импорт бибилиотек
import pandas as pd
import seaborn as sns
# вызываем из sklearn DBSCAN
from sklearn.cluster import DBSCAN
```

```python
# load data
df = pd.read_csv('data/pokemon.csv')
# произведем кластеризацию покемонов по уровню атаки и защиты
X = df[['Attack', 'Defense']]

# визуализируем через точечную диаграмму по двум признакам
sns.scatterplot(x=X.Attack, y=X.Defense)
```

```python
X
```

```python
# Импортируем DBSCAN-кластеризацию
dbscan_clustering = DBSCAN(eps=3, min_samples=3)

# Запускаем кластеризацию на наших данных
dbscan_clustering.fit(X)

# Далее можно визуализировать результаты, как мы делали это с алгоритмом k-means
sns.scatterplot(x=X.Attack, y=X.Defense, c=dbscan_clustering.labels_)

print('Алгоритм нашел выброс и не включил его ни в один кластер')
```

Задание 7.5 <br>
Визуализируйте результаты. Сколько точек было размечено как выбросы?

```python
# load data
X = pd.read_csv('data/housing.csv')
X.head()
```

```python
# Импортируем DBSCAN-кластеризацию
dbscan_housing = DBSCAN()

# Запускаем кластеризацию на наших данных
dbscan_housing.fit(X)

# Далее можно визуализировать результаты, как мы делали это с алгоритмом k-means
sns.scatterplot(x=X.year, y=X.price, c=dbscan_housing.labels_)

print('При значениях по умолчанию алгоритм нашел 2 выброса')
```

```python
# Импортируем DBSCAN-кластеризацию
dbscan_housing = DBSCAN(eps = 1.15, min_samples = 2)

# Запускаем кластеризацию на наших данных
dbscan_housing.fit(X)

# Далее можно визуализировать результаты, как мы делали это с алгоритмом k-means
sns.scatterplot(x=X.year, y=X.price, c=dbscan_housing.labels_)

print('При значениях по умолчанию алгоритм нашел 2 выброса')
```
